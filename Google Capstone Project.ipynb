```python
#Converting to Date time format column

activity["ActivityDate"]=pd.to_datetime(activity["ActivityDate"])
```

```python
#Calculating Day of the Week

activity["Day of the Week"]=activity["ActivityDate"].dt.day_name()
```

```python
#Calculating total minutes

activity["total_mins"] = activity["VeryActiveMinutes"] + activity["FairlyActiveMinutes"] + activity["LightlyActiveMinutes"] + activity["SedentaryMinutes"]
```

```python
#Merging two Datasets

df=pd.merge(activity,Weight,on='Id',how='outer')
```

```python
#Data Visualization

sns.scatterplot(data=df,x="TotalSteps",y="Calories")
plt.title("No of Calories Burnt Vs Total Steps")
```

```python
plt.hist(df["Day of the Week"],bins=7,width=0.6)
```

```python
very_active_mins = df["VeryActiveMinutes"].sum()
fairly_active_mins = df["FairlyActiveMinutes"].sum()
lightly_active_mins = df["LightlyActiveMinutes"].sum()
sedentary_mins = df["SedentaryMinutes"].sum()
```

```python
minutes=[very_active_mins, fairly_active_mins, lightly_active_mins, sedentary_mins]
label = ["Very active minutes", "Fairly active minutes", "Lightly active minutes", "Sedentary minutes"]
```

```python
plt.pie(minutes,autopct='%1.1f',labels=label)
plt.title("Percentage of Each Activity in Minutes")
```
